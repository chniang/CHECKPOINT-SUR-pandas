```python
import pandas as pd
import numpy as np

# Données du dictionnaire
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'tentatives': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualifier': ['oui', 'non', 'oui', 'non', 'non', 'oui', 'oui', 'non', 'non', 'oui']
}

# Étiquettes d'index
etiquettes = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Création du DataFrame
df = pd.DataFrame(exam_data, index=etiquettes)

# Affichage du DataFrame
print(df)
# Affichage des trois premières lignes du DataFrame
print(df.head(3))
# Suppression des lignes contenant des valeurs NaN
nouveau_df = df.dropna()

# Affichage du DataFrame nettoyé
print(nouveau_df)
# Extraction des colonnes 'name' et 'score'
df_extrait = df.loc[:, ['name', 'score']]

# Affichage du DataFrame extrait
print(df_extrait)
# Ajout d'une nouvelle ligne 'k' au DataFrame
df.loc['k'] = ['Suresh', 15.5, 1, 'oui']

# Affichage du DataFrame mis à jour
print(df)
# Suppression de la colonne 'tentatives'
df = df.drop(columns=['tentatives'])

# Affichage du DataFrame mis à jour
print(df)
# Ajout de la colonne 'Succès'
df['Succès'] = df['score'].apply(lambda x: 1 if x > 10 else 0)

# Affichage du DataFrame mis à jour
print(df)




```

```python
# Exportation du DataFrame dans un fichier CSV
df.to_csv('my_data.csv', index=True)

# Confirmation de l'exportation
print("Le DataFrame a été exporté avec succès dans le fichier 'my_data.csv'.")
```
